```python
import geopandas
data = geopandas.read_file(input_filename)
type(data)
```

```python
# Select columns
data = data[["RYHMA", "LUOKKA", "geometry"]]

#Rename columns
data = data.rename(
    columns={
        "RYHMA": "GROUP",
        "LUOKKA": "CLASS"
    }
)
```

### Map projections

```python
# Import data from the GISCO
import pathlib

MAIN_PATH = pathlib.Path().resolve()
NOTEBOOK_PATH = pathlib.Path().resolve() / "notebooks" / "lessons" / "lesson-2"
DATA_DIRECTORY = NOTEBOOK_PATH / "data"
```

```python
import geopandas

eu_countries = geopandas.read_file(
    DATA_DIRECTORY / "eu_countries" / "eu_countries_2022.gpkg"
)
```

```python
# Check the CRS information from the data
eu_countries.crs
```

```python
eu_countries.geometry.head()
```

```python
# Transforming a EPSG 4326 to a 3035, a projection with less distortion for the EU countries
eu_countries_EPSG3035 = eu_countries.to_crs("EPSG:3035")
```

```python
eu_countries_EPSG3035.geometry.head()
```

```python
# Represent both representations
import matplotlib.pyplot

# Create a image with subplots, so the images are side by side
figure, (axis1, axis2) = matplotlib.pyplot.subplots(nrows=1, ncols=2)

#Plot the 4326 data
eu_countries.plot(ax=axis1)
axis1.set_title("WGS84")
axis1.set_aspect(1)

#Plot the ESPG 3035 data
eu_countries_EPSG3035.plot(ax=axis2)
axis2.set_title("ETRS-LAEA")
axis2.set_aspect(1)

matplotlib.pyplot.tight_layout()
```

```python
eu_countries_EPSG3035.to_file(
    MAIN_PATH / "lesson-2" / "eu_countries" / "eu_countries_EPSG3035.gpkg"
)
```

### Global map projections

```python
# EXTRA Exercise
import pathlib
import geopandas

# import file
world_countries = geopandas.read_file(DATA_DIRECTORY / "world_countries" / "ne_110m_admin_0_countries.zip")

# check CRS
print(world_countries.crs)

# Change projections
world_countries_EPSG3035 = world_countries.to_crs("EPSG:3035")
world_countries_other = world_countries.to_crs("EPSG:3857")

# Plot the three projections
import matplotlib.pyplot

# Create a image with subplots, so the images are side by side
figure, (axis1, axis2, axis3) = matplotlib.pyplot.subplots(nrows=1, ncols=3)

#Plot the 4326 data
world_countries.plot(ax=axis1)
axis1.set_title(f"{world_countries.crs.name}")
axis1.set_aspect(1)

#Plot the ESPG 3035 data
world_countries_EPSG3035.plot(ax=axis2)
axis2.set_title(f"{world_countries_EPSG3035.crs.name}")
axis2.set_aspect(1)

#Plot the Robinson CRS
world_countries_other.plot(ax=axis3)
axis2.set_title(f"{world_countries_other.crs.name}")
axis2.set_aspect(1)

matplotlib.pyplot.tight_layout()
```

```python
# Eckert-IV (https://spatialreference.org/ref/esri/54012/)
ECKERT_IV = "+proj=eck4 +lon_0=0 +x_0=0 +y_0=0 +ellps=WGS84 +datum=WGS84 +units=m +no_defs"

world_countries_eckert_iv = world_countries.to_crs(ECKERT_IV)
world_countries_eckert_iv.plot()
matplotlib.pyplot.title("Eckert Ⅳ")
matplotlib.pyplot.axis("off")
```

```python
# An orthographic projection, centered in Finland!
# (http://www.statsmapsnpix.com/2019/09/globe-projections-and-insets-in-qgis.html)
world_countries_ortho = world_countries.to_crs(
    "+proj=ortho +lat_0=60.00 +lon_0=23.0000 +x_0=0 +y_0=0 "
    "+a=6370997 +b=6370997 +units=m +no_defs"
)
world_countries_ortho.plot()
matplotlib.pyplot.axis("off")
```
